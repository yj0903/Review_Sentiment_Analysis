### sns 데이터

```python
import os
import json  
import pandas as pd

# './data' 경로 설정
data_path = './Sample/02.라벨링데이터/SNS/01. 패션'

# './data' 경로 안에 있는 모든 JSON 파일 리스트 가져오기
json_files = [f for f in os.listdir(data_path) if f.endswith('.json')]

# 각 JSON 파일을 불러와 DataFrame으로 추가
dfs = []
for json_file in json_files:
    file_path = os.path.join(data_path, json_file)
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
        dfs.append(pd.json_normalize(data, sep='_'))

# 모든 DataFrame을 합치기
result_df = pd.concat(dfs, ignore_index=True)

# 결과 출력
result_df
```

### 쇼핑몰 데이터

```python
import os
import json  
import pandas as pd


# './data' 경로 설정
name = '01. 패션'
data_path = './Sample/02.라벨링데이터/쇼핑몰/'+ name

# './data' 경로 안에 있는 모든 폴더 리스트 가져오기
subdirectories = [d for d in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, d))]

# 모든 폴더 안의 JSON 파일을 불러와 DataFrame으로 추가
dfs = []
for subdir in subdirectories:
    subdir_path = os.path.join(data_path, subdir)
    json_files = [f for f in os.listdir(subdir_path) if f.endswith('.json')]
    
    for json_file in json_files:
        file_path = os.path.join(subdir_path, json_file)
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
            dfs.append(pd.json_normalize(data, sep='_'))

# 모든 DataFrame을 합치기
result_df = pd.concat(dfs, ignore_index=True)
result_df
```

```python
# 결과 출력
result_df.to_csv('./data/'+name+'.csv', header = True, index = False, encoding='utf-8-sig')
```
